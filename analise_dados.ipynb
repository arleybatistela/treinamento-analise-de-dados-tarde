Criar duas listas:

- 5 produtos (apenas inserir valores).
- A primeira lista, custo do produto.
- A segunda lista, o preço de venda.
- Desafio: Calcular o lucro de cada produto.

```python
custo_produtos = [50,100,150,200,250]
preco_venda = [100,200,300,400,500]

array_custo = np.array(custo_produtos)
array_preco = np.array(preco_venda)

lucro_produto = array_preco - array_custo
print(lucro_produto)
print(type(lucro_produto))
```

```python
salarios = np.array([2,3,4,5,6,500])
maior_salario = np.max(salarios)
print(maior_salario)
```

```python
media_salarios = np.mean(salarios)
print(f"{media_salarios:.2f}")
```

```python
menor_salario = np.min(salarios)
print(menor_salario)
```

```python
mediana_salarios = np.median(salarios)
print(mediana_salarios)
```

```python
desvio = np.std(salarios)
print(desvio)
```
