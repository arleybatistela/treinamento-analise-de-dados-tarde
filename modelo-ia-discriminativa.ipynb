```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r".\data\dataset_studante_tratado.csv")
sns.lmplot(data=df,x="study_hours_per_day",y="exam_score");
```

Separando dados de entrada com os dados que queremos descobrir.

```python
X = df[["study_hours_per_day","social_media_hours","netflix_hours","attendance_percentage","sleep_hours","exercise_frequency","mental_health_rating"]]
y = df["exam_score"]
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(X,y)
```

```python
novo_aluno = pd.DataFrame({"study_hours_per_day":[6],"social_media_hours":[6],"netflix_hours":[6],"attendance_percentage":[90],"sleep_hours":[6],"exercise_frequency":[6],"mental_health_rating":[6]})
```

```python
resultado = modelo.predict(novo_aluno)
print(resultado)
```

```python
# Coeficiente angular
modelo.coef_
```

```python
# Coeficiente linaar
modelo.intercept_
```

```python
modelo.score(X,y)
```
