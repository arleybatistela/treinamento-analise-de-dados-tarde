```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r"data\dataset_aula01_techshop_vendas.csv")
df.info()
```

```python
sns.countplot(data=df,x="categoria")
```

```python
grafico = sns.countplot(data=df,x="categoria");
grafico.set_title("Contagem por categoria");
grafico.set_ylabel("Quantidade");
grafico.set_xlabel("categoria");
```

```python
grafico = sns.histplot(data=df,x="preco_unitario",kde=True,bins=20);
grafico.set_title("Histograma");
grafico.set_ylabel("Quantidade");
grafico.set_xlabel("preço_unitario");
```

```python
grafico = sns.boxplot(data=df,x="categoria",y="preco_unitario");
grafico.set_title("Distribuição de preços por categoria");
grafico.set_ylabel("Preço Unitário");
grafico.set_xlabel("Categoria");
```
